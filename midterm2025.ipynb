# หาคนบงการ

จากคำให้การของฆาตกร ที่บอกว่าถูกว่าจ้างโดยผู้หญิงที่มีเงินมาก อาจจะเกี่ยวข้องกับรายได้

```python
income = '/content/drive/MyDrive/Python&DataViz2025/Data Midterm/income.csv'
income = pd.read_csv(income)
income
```

```python
person = '/content/drive/MyDrive/Python&DataViz2025/Data Midterm/person.csv'
person = pd.read_csv(person)
person
```

```python
# เชื่อมตาราง person กับ income ด้วย ssn
merged_person_income_df = pd.merge(person, income, on='ssn', how='inner')
display(merged_person_income_df)
```

```python
# เรียงลำดับตาราง merged_person_income_df จากรายได้มากไปน้อย
merged_person_income_sorted = merged_person_income_df.sort_values(by='annual_income', ascending=False)
display(merged_person_income_sorted)
```

ตรวจสอบเพศของคนที่มีรายได้มากว่าเป็นผู้หญิงหรือไม่

```python
drivers_license = '/content/drive/MyDrive/Python&DataViz2025/Data Midterm/drivers_license.csv'
drivers_license = pd.read_csv(drivers_license)
drivers_license
```

```python
# เชื่อมตาราง person และ drivers_license โดยใช้ license_id จาก person และ id จาก drivers_license
merged_person_license_df = pd.merge(person, drivers_license, left_on='license_id', right_on='id', how='inner')
display(merged_person_license_df)
```

```python
# นำข้อมูล annual_income มาต่อเข้ากับ merged_person_license_df โดยใช้ ssn
final_merged_df = pd.merge(merged_person_license_df, merged_person_income_df[['ssn', 'annual_income']], on='ssn', how='left')
display(final_merged_df)
```

```python
# เรียงข้อมูล final_merged_df ตาม annual_income จากมากไปน้อย
final_merged_sorted = final_merged_df.sort_values(by='annual_income', ascending=False)
display(final_merged_sorted)
```

```python
# กรองข้อมูลเพื่อหาบุคคลที่เป็นผู้หญิง
female_individuals = final_merged_sorted[final_merged_sorted['gender'] == 'female']
display(female_individuals)
```
