# Quiz11

plot กราฟ 3 อันในแนวทแยงมุม

1.   boxplot
2.   scatter
3.  histogram



````
x = stopandsearch_df['Latitude'].dropna()
y = stopandsearch_df['Longitude'].dropna()

# สร้าง subplot 3x3 แต่ใช้เฉพาะตำแหน่งแนวทแยงมุม
fig, axes = plt.subplots(3, 3, figsize=(10, 10))

# ลบแกนที่ไม่ได้ใช้
for i in range(3):
    for j in range(3):
        if i != j:
            axes[i, j].axis("off")

# กราฟ 1: Boxplot (ซ้ายบน)
axes[0, 0].boxplot(x)
axes[0, 0].set_title("Boxplot")

# กราฟ 2: Scatter (ตรงกลาง)
axes[1, 1].scatter(x, y, alpha=0.7)
axes[1, 1].set_title("Scatter Plot")

# กราฟ 3: Histogram (ขวาล่าง)
axes[2, 2].hist(x, bins=15, color="skyblue", edgecolor="black")
axes[2, 2].set_title("Histogram")

plt.tight_layout()
plt.show()
````
