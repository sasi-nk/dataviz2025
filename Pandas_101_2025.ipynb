````
import pandas as pd
import glob
import os

# 1. ทำการเชื่อมต่อกับ Google Drive (เหมือนเดิม)
from google.colab import drive
drive.mount('/content/drive')

# 2. กำหนด Path หลักและรูปแบบของไฟล์ที่ต้องการค้นหา
# เราจะใช้เครื่องหมาย * (Wildcard) แทนส่วนของชื่อไฟล์หรือโฟลเดอร์ที่เปลี่ยนแปลงไป
base_path = '/content/drive/MyDrive/Python&DataViz2025/Data/'
# รูปแบบ: ค้นหาไฟล์ .csv ที่อยู่ในโฟลเดอร์ 2025-XX ใดๆ ก็ตาม
path_pattern = f"{base_path}2025-*/*-stop-and-search.csv"
all_stopandsearch = glob.glob(path_pattern)

print(f"ค้นพบไฟล์ CSV ทั้งหมด: {len(all_stopandsearch)} ไฟล์")
# ลองแสดงตัวอย่าง 5 ไฟล์แรกที่เจอ
print("ตัวอย่างไฟล์ที่พบ:")
for f in all_stopandsearch[:5]:
    print(f)
print("-" * 50)

# 4. สร้าง Loop เพื่ออ่านไฟล์ทั้งหมดและรวมเป็น DataFrame เดียว
all_dfs = [] # สร้าง list ว่างเพื่อเก็บ DataFrame ชั่วคราว

for filepath in all_stopandsearch:
    # อ่านไฟล์ CSV จาก path ที่ได้มา
    df = pd.read_csv(filepath)

    # --- ส่วนสำคัญ: ดึงชื่อเขตตำรวจจากชื่อไฟล์ ---
    # ชื่อไฟล์: '.../2025-01-avon-and-somerset-stop-and-search.csv'
    # 1. ดึงเฉพาะชื่อไฟล์ออกมา: os.path.basename(filepath) -> '2025-01-avon-and-somerset-stop-and-search.csv'
    filename = os.path.basename(filepath)
    # 2. แยกส่วนประกอบของชื่อไฟล์ด้วยเครื่องหมาย '-'
    parts = filename.split('-')
    # 3. ชื่อเขตตำรวจจะอยู่ระหว่างส่วนของวันที่ (2 ส่วนแรก) และ 3 ส่วนท้าย ('stop', 'and', 'search.csv')
    force_name_parts = parts[2:-3]
    # 4. นำชิ้นส่วนชื่อมารวมกันและจัดรูปแบบตัวพิมพ์ให้สวยงาม (เช่น 'avon and somerset' -> 'Avon And Somerset')
    force_name = ' '.join(force_name_parts).title()

    # เพิ่มคอลัมน์ 'Police Force' เข้าไปใน DataFrame
    df['Police Force'] = force_name

    all_dfs.append(df)

# รวม DataFrame ทั้งหมดใน list ให้เป็น DataFrame เดียว
combined_df = pd.concat(all_dfs, ignore_index=True)

print("รวมข้อมูลจากทุกไฟล์สำเร็จแล้ว!")
print(f"DataFrame ใหม่มีข้อมูลทั้งหมด: {len(combined_df):,} แถว")
print(f"จากเขตตำรวจทั้งหมด: {combined_df['Police Force'].nunique()} เขต")
#print("-" * 50)

# --- จากตรงนี้ สามารถนำโค้ดวิเคราะห์จากคำตอบที่แล้วมาใช้ต่อได้เลย ---

# คัดกรองข้อมูลเฉพาะผู้ถูกเรียกค้นที่เป็นคน Asia
asian_stops_df = combined_df[combined_df['Self-defined ethnicity'].str.contains('Asian', na=False)].copy()

# จัดกลุ่มและนับจำนวนตามเขตตำรวจ
asian_counts_by_force = asian_stops_df.groupby('Police Force').size()
asian_counts_sorted = asian_counts_by_force.rename('Total Asian Stops').sort_values(ascending=False)

# หาค่าและแสดงผลลัพธ์
most_stops_force = asian_counts_sorted.idxmax()
most_stops_count = asian_counts_sorted.max()

least_stops_force = asian_counts_sorted.idxmin()
least_stops_count = asian_counts_sorted.min()

print("สรุปผลการเรียกค้นบุคคลเชื้อสายเอเชียในรอบ 5 เดือน (จากข้อมูลที่โหลดแบบอัตโนมัติ):\n")
print(f"📈 เขตที่มีการเรียกค้นมากที่สุด: {most_stops_force} ({most_stops_count:,} ครั้ง)")
print(f"📉 เขตที่มีการเรียกค้นน้อยที่สุด: {least_stops_force} ({least_stops_count:,} ครั้ง)")
````

- วนลูปนับว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง

- มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน

- คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่ หรือ พ่อ มากที่สุด

````
# prompt: จาก data this_data วนลูปนับว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง

count = 0
stories = []
for index, row in this_data.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    count += 1
    stories.append(row['title'])

print(f"There are {count} stories with 'แม่' or 'พ่อ' in the title.")
print("The stories are:")
for story in stories:
  print(story)
````

````
# prompt: มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน บอกจำนวนด้วย

import pandas
import pandas as pd
import os
from google.colab import drive
from bs4 import BeautifulSoup

# ... (preceding code)

count_life = 0
count_local = 0
for index, row in this_data.iterrows():
  if ('แม่' in row['title'] or 'พ่อ' in row['title']) and row['type'] == 'นิทานชีวิต':
    count_life += 1
  elif ('แม่' in row['title'] or 'พ่อ' in row['title']) and row['type'] == 'นิทานประจำถิ่น':
    count_local += 1

if count_life > count_local:
  print(f"There are more stories with 'แม่' or 'พ่อ' in the title in นิทานชีวิต ({count_life} stories) than in นิทานประจำถิ่น ({count_local} stories).")
elif count_local > count_life:
  print(f"There are more stories with 'แม่' or 'พ่อ' in the title in นิทานประจำถิ่น ({count_local} stories) than in นิทานชีวิต ({count_life} stories).")
else:
  print(f"There are an equal number of stories with 'แม่' or 'พ่อ' in the title in both นิทานชีวิต and นิทานประจำถิ่น ({count_life} stories each).")
````

````
# prompt: คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่ หรือ พ่อ มากที่สุด (พร้อมทั้งอธิบาย code แต่ละบรรทัดด้วย)

# สร้าง dictionary เพื่อเก็บจำนวนนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องสำหรับแต่ละคนแต่ง
creator_counts = {}
for index, row in this_data.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    creator = row['creator']
    if creator not in creator_counts:
      creator_counts[creator] = 0
    creator_counts[creator] += 1

# หาคนแต่งที่มีจำนวนนิทานมากที่สุด
max_count = 0
most_prolific_creator = None
for creator, count in creator_counts.items():
  if count > max_count:
    max_count = count
    most_prolific_creator = creator

print(f"The creator who wrote the most stories with 'แม่' or 'พ่อ' in the title is {most_prolific_creator} with {max_count} stories.")
````
