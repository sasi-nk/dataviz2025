# pandas: parallel coordinate

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df.head()
```

```python
matplotlib.rcParams['figure.figsize'] = (5, 5) # กำหนดขนาดของ figure

pd.plotting.parallel_coordinates(iris_df, 'target')
```

อาจจะสื่อสารได้ดีกว่าภาพแรก แต่ข้อจำกัดคือ เดต้าต้องน้อยๆ

```python
pd.plotting.scatter_matrix(iris_df);
```

scatter คือ กระจัดกระจาย

เส้นทแยงมุมที่เป็นกราฟฮิสโตแกรมเพราะ มีความสัมพันธ์ในตัวมันเอง

# Color

```python
matplotlib.rcParams['figure.figsize'] = (15, 15) # กำหนดขนาดของ figure

plt.imshow(iris_df.iloc[:,:], interpolation='nearest', cmap='Pastel2')
```

```python
matplotlib.rcParams['figure.figsize'] = (5, 5) # กำหนดขนาดของ figure

pd.plotting.parallel_coordinates(iris_df, 'target', color=['orange', 'green', 'blue'])
```

```python
matplotlib.rcParams['figure.figsize'] = (5, 5) # กำหนดขนาดของ figure

pd.plotting.parallel_coordinates(iris_df, 'target', color=['c', 'm', 'y']); # ตัวย่อชื่อสี c m y k r g b w
```

```python
matplotlib.rcParams['figure.figsize'] = (5, 5) # กำหนดขนาดของ figure

pd.plotting.parallel_coordinates(iris_df, 'target', color=[[1,0,0], [0.5,0.5,0.5], [0,1,1]]); # ค่าสี R G B
```

```python
matplotlib.rcParams['figure.figsize'] = (5, 5) # กำหนดขนาดของ figure

pd.plotting.parallel_coordinates(iris_df, 'target', color=['#F23362', '#337CF2','#F29F33' ]); # html color code
```
